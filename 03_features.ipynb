```python
import pandas as pd
import numpy as np
```

```python
visits = spark.table("workspace.default.air_visits").toPandas()
stores = spark.table("workspace.default.air_stores").toPandas()
dates = spark.table("workspace.default.date_info").toPandas()
```

```python
visits["visit_date"] = pd.to_datetime(visits["visit_date"])
dates["calendar_date"] = pd.to_datetime(dates["calendar_date"])
```

```python
df = visits.merge(dates, left_on="visit_date", right_on="calendar_date", how="left")
df = df.merge(stores, on="air_store_id", how="left")
```

```python
df.head()
```

```python
df = df.sort_values(["air_store_id", "visit_date"]).reset_index(drop=True)
```

```python
df.head()
```

```python
df["month"] = df["visit_date"].dt.month
df.head()
```

```python
df['is_weekend'] = df["day_of_week"].isin(["Saturday", "Sunday"]).astype(int)
df.head()
```

```python
for k in [1, 7, 14, 28]:
    df[f"lag_{k}"] = df.groupby("air_store_id")["visitors"].shift(k)
```

```python
df.sample(5)
```

```python
for w in [7, 28]:
    df[f"roll_mean_{w}"] = (
        df.groupby("air_store_id")["visitors"]
          .transform(lambda s: s.shift(1).rolling(w).mean())
    )
    df[f"roll_std_{w}"] = (
        df.groupby("air_store_id")["visitors"]
          .transform(lambda s: s.shift(1).rolling(w).std())
    )
```

```python
df.sample(5)
```

```python
np.log(df["visitors"])
```

```python
df["y"] = np.log1p(df["visitors"])
```

```python
df.head()
```

```python
df.count()
```

```python
engineered = ["lag_1","lag_7","lag_14","lag_28",
              "roll_mean_7","roll_std_7","roll_mean_28","roll_std_28"]

df_model = df.dropna(subset=engineered).reset_index(drop=True)
```

```python
df_model.count()
```

```python
df_model.info()
```

```python
spark.createDataFrame(df_model).write.format("delta").mode("overwrite") \
    .saveAsTable("workspace.default.features")
```
