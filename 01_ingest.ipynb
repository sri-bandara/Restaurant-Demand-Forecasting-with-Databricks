```python
visits_path = "/Volumes/workspace/default/raw/air_visit_data.csv"

visits = spark.read.csv(visits_path, header=True, inferSchema=True)
visits.display()
```

```python
visits.printSchema()
```

```python
visits.write.format("delta").mode("overwrite").saveAsTable("workspace.default.air_visits")
```

```python
spark.sql("SELECT COUNT(*) AS rows FROM workspace.default.air_visits").display()
```
